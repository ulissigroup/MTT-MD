```python
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize, signal
from lmfit import models
from scipy import stats
import seaborn as sns

def generate_model(spec):
    composite_model = None
    params = None
    x = spec['x']
    y = spec['y']
    x_min = np.min(x)
    x_max = np.max(x)
    x_range = x_max - x_min
    y_max = np.max(y)
    for i, basis_func in enumerate(spec['model']):
        prefix = f'm{i}_'
        model = getattr(models, basis_func['type'])(prefix=prefix)
        if basis_func['type'] in ['GaussianModel', 'LorentzianModel', 'VoigtModel']: # for now VoigtModel has gamma constrained to sigma
            model.set_param_hint('sigma', min=1e-6, max=x_range)
            model.set_param_hint('center', min=x_min, max=x_max)
            model.set_param_hint('height', min=1e-6, max=1.1*y_max)
            model.set_param_hint('amplitude', min=1e-6)
            # default guess is horrible!! do not use guess()
            default_params = {
                prefix+'center': x_min + x_range * random.random(),
                prefix+'height': y_max * random.random(),
                prefix+'sigma': x_range * random.random()
            }
        else:
            raise NotImplemented(f'model {basis_func["type"]} not implemented yet')
        if 'help' in basis_func:  # allow override of settings in parameter
            for param, options in basis_func['help'].items():
                model.set_param_hint(param, **options)
        model_params = model.make_params(**default_params, **basis_func.get('params', {}))
        if params is None:
            params = model_params
        else:
            params.update(model_params)
        if composite_model is None:
            composite_model = model
        else:
            composite_model = composite_model + model
    return composite_model, params
                                 
def update_spec_from_peaks(spec, model_indicies, peak_widths=(10, 25), **kwargs):
    x = spec['x']
    y = spec['y']
    x_range = np.max(x) - np.min(x)
    peak_indicies = signal.find_peaks_cwt(y, peak_widths)
    np.random.shuffle(peak_indicies)
    for peak_indicie, model_indicie in zip(peak_indicies.tolist(), model_indicies):
        model = spec['model'][model_indicie]
        if model['type'] in ['GaussianModel', 'LorentzianModel', 'VoigtModel']:
            params = {
                'height': y[peak_indicie],
                'sigma': x_range / len(x) * np.min(peak_widths),
                'center': x[peak_indicie]
            }
            if 'params' in model:
                model.update(params)
            else:
                model['params'] = params
        else:
            raise NotImplemented(f'model {basis_func["type"]} not implemented yet')
    return peak_indicies
                                 
def print_best_values(spec, output):
#     print('center'+ '\t' +'\t' +'\t' + 'sigma' + '\t' +'\t' +'\t' + 'height')
    centers = []
    sigmas = []
    heights = []
    counts = []
    components = output.eval_components(x=spec['x'])                             
    for i in range(len(spec['model'])):
        center = output.best_values['m%d_center'%i]
        sigma = output.best_values['m%d_sigma'%i]
        amp = output.best_values['m%d_amplitude'%i]
        height = amp/(np.sqrt(2*np.pi)*sigma)
        count = components['m%d_'%i]
        counts += [sum(count)]
        centers += [center]
        sigmas += [sigma]
        heights += [height]
#         print(center, sigma, height)
    return np.array(centers), np.array(sigmas), np.array(heights), np.array(counts)
```

```python
# Works well for 2 gaussians but not for 3 gaussians (eg. [1, 2])
def get_angle(in_list, in_moment, is_log, peak_width):
    df_counts = []
    angles_counts = []
    angles = []
    centers = []
    sigmas = []
    heights = []
    sum_counts = []
    for i in range(len(in_list)):
        nan_idx = np.where(~np.isnan(in_list[i]))[0]
        in_list[i] = np.array(in_list[i])[nan_idx]

        hist, bin_edge= np.histogram(in_list[i], bins='auto')
        if is_log:
            xs = np.where(bin_edge > 0.1)[0][:-1] # Only consider -1 < log(angle) < 2
            hist = hist[xs]
            bin_edge = bin_edge[xs]
        else:
            xs = np.arange(len(bin_edge))[:-1]
            hist = hist[xs]
            bin_edge = bin_edge[xs]
            
        spec = {
            'x': bin_edge,
            'y': hist,
            'model': [
                {'type': 'GaussianModel'}, # Assume the distribution is Gaussian
            ]
        }
        peaks_found = update_spec_from_peaks(spec, np.arange(len(spec['model'])), peak_widths=(peak_width,))
    # To make sure # of gaussian = # peaks found
        if len(spec['model']) != len(peaks_found):
            spec['model'] = []
            for j in range(len(peaks_found)):
                spec['model'].append({'type': 'GaussianModel'})
            peaks_found = update_spec_from_peaks(spec, np.arange(len(spec['model'])), peak_widths=(peak_width,))
    #     print('peaks at', bin_edge[peaks_found])

        if len(peaks_found) == 3 or i == 11 :
            model = models.GaussianModel(prefix = 'm0_')
            for n in range(1, len(peaks_found)):
                model += models.GaussianModel(prefix = 'm%d_' %n)
                output = model.fit(spec['y'], x=spec['x'])
        else:
            model, params = generate_model(spec)
            output = model.fit(spec['y'], params, x=spec['x'])
            
#         fig, gridspec = output.plot(data_kws={'markersize':  1})
#         fig, ax = plt.subplots()
#         ax.scatter(spec['x'], spec['y'], s=4)
#         components = output.eval_components(x=spec['x'])
#         for i, model in enumerate(spec['model']):
#             ax.plot(spec['x'], components[f'm{i}_'])

#         for i in peaks_found:
#             ax.axvline(x=spec['x'][i], c='black', linestyle='dotted')
        
        counts = pd.DataFrame(data = output.best_fit, columns = ['counts'])
#         angles = pd.DataFrame(data = spec['x'], columns = ['log_angles'])
#         moments = pd.DataFrame(data = in_moment[i][xs], columns = ['Moments'])
        center, sigma, height, sum_count = print_best_values(spec,output)
        idx = np.argsort(center)
        centers.append(center[idx])
        sigmas.append(sigma[idx])
        sum_counts.append(sum_count[idx])
#         heights.append(height[idx]/len(in_list[i]))
        print(sum_count)
#         angles_counts.append(pd.concat([moments, angles, counts], axis = 1))
#     angles_counts = pd.concat(angles_counts)
    centers_val, counts_all, counts_cache = get_values(centers, sigmas, sum_counts, heights)
    
    return centers_val, counts_all, counts_cache


def get_height(in_list):
    centers = []
    heights = []
    for i in range(len(in_list)):
        if i >= 6:
            peak_width = 10
        elif i == 12:
            peak_width = 10
        elif i == 0:
            peak_width = 1.5
        elif i == 1:
            peak_width = 2
        elif i < 6:
            peak_width = 1.8

        x = in_list[i]
        nan_idx = np.where(~np.isnan(x))[0]
        x = np.array(x)[nan_idx]

        hist, bin_edge= np.histogram(x, bins='auto')
        xs = np.arange(len(bin_edge))[:-1]
        hist = hist[xs]
        bin_edge = bin_edge[xs]
#         model_1 = models.GaussianModel(prefix='m0_')
#         model_2 = models.GaussianModel(prefix='m1_')
#         model = model_1 + model_2

#         params_1 = model_1.make_params(center=1, sigma=1)
#         params_2 = model_2.make_params(center=-1, sigma=1)
#         params = params_1.update(params_2)

#         output = model.fit(hist, params, x=bin_edge)
#         fig, gridspec = output.plot(data_kws={'markersize': 1})

        peakind = signal.find_peaks(hist, np.arange(len(bin_edge)), width= peak_width)
        # print(peakind)
        centers.append(bin_edge[peakind[0][:2]])
        if len(peakind[0]) > 2:
            above_mono = np.sum(peakind[1]["width_heights"][1:])
            below_mono = peakind[1]["width_heights"][0]

            heights.append((above_mono, below_mono)/(np.sum(peakind[1]["width_heights"])))

        elif len(peakind[0]) == 2:
            heights.append(peakind[1]["width_heights"][:2]/(np.sum(peakind[1]["width_heights"][:2])))
    return centers, heights

```
