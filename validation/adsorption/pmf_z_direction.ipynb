```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import math
matplotlib.rcParams.update({'font.size': 22})
plt.rc('figure', figsize=(11,6))


with open('./graphene_ads/output/cntabf.pmf') as f:
    pmf = []
    for line in f:
        if line[0] != '#':
            pmf.append(line.split())
    pmf.pop(-1)
    
    newpmf = []
    integer = []
    for item in pmf:
        for number in item:
            integer.append(float(number))
        newpmf.append(integer)
        integer = []
        zippedPmf = newpmf[1:]    
    distance = np.array(zippedPmf)[:,0]
    energy = np.array(zippedPmf)[:,1]


plt.plot(np.array(distance)*0.1, np.array(energy)-energy[-1])
# plt.scatter(np.array(distance)*0.1, np.array(energy)-energy[-1])
plt.xlabel('Height above the surface (nm)')
plt.ylabel('Free Energy (Kcal/mol)')
plt.ylim(-5,3)
plt.savefig('pmf')
plt.show()
```
